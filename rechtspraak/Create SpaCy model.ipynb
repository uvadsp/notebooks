### Create word lists

```python
def create_word2vec_relevant_words(words, matches):
    word2vec_list = []
    for word in words:
        results = dutch_word2vec_model.wv.most_similar(word, topn=100)
        for i in results:
            word2vec_list.append(i[0])
            
    word2vec_list = list(set([i for i in word2vec_list if word2vec_list.count(i)>matches]))
    return word2vec_list
```

```python
list_of_drugs = ['xtc', 'mdma', 'cocaine', 'wiet', 'speed', 'bmk', 'pmk']
word2vec_drug_list = create_word2vec_relevant_words(list_of_drugs, 2)

print(len(word2vec_drug_list))
print(word2vec_drug_list)
```

### Custom edit country dict

```python
for key in country_translation_dict:
    if country_translation_dict[key] == None:
        country_translation_dict[key] = []
edit_dict = {'Netherlands': ['bergen op zoom', 'gessel', 'montfort', 'kiezebrink', 'bilt'],
            'unknown': ['buitenland'],
            'Russia': ['rusland'],
            'United Kingdom': ['birmingham']}
remove_list = ['-3054', 'A14', '.174', 'berg', '.394', 'joh', 'face-to-face', 'doos/box', 'café/bar/discotheek', 'engelsman']

for key in edit_dict:
    for loc in edit_dict[key]:
        if loc not in remove_list:
            remove_list.append(loc)

## Remove unwanted
for key in country_translation_dict:
    for loc in remove_list:
        if loc in country_translation_dict[key]:
            country_translation_dict[key].remove(loc)

## Add wanted
for key in edit_dict:
    for loc in edit_dict[key]:
        if key not in country_translation_dict:
            country_translation_dict[key] = []
        if loc not in country_translation_dict[key]:
            country_translation_dict[key].append(loc)
```

```python
nlp = spacy.load('nl_core_news_lg')
# nlp = spacy.load('nl_core_news_md')
```

```python
def configure_spacy_model():
    # Create dict of drug pattern and quantity pattern
    pattern_list = []
    for i in drugs_list:
        pattern_list.append({"label": "DRUG", "pattern": [{"lower": i.lower()}]})
    
    for i in word2vec_quantity_list:
        pattern_list.append({"label": "QUANTITY", "pattern": [{"IS_DIGIT": True}, {"LOWER": i}]})
        pattern_list.append({"label": "QUANTITY", "pattern": [{"ENT_TYPE": "CARDINAL"}, {"LOWER": i}]})
    
    
    
    for i in countries_to_exclude:
        word_list = i.strip().split(' ')
        pattern = []
        for word in word_list:
            pattern.append({"lower": word})
        pattern_list.append({"label": "EXCL", "pattern": pattern})
        
#     for i in countries_list:
#         pattern_list.append({"label": "GPE", "pattern": {"lower": i.strip().lower()}})
        
    for i in countries_list:
        word_list = i.strip().split(' ')
        pattern = []
        for word in word_list:
            pattern.append({"lower": word})
        pattern_list.append({"label": "GPE", "pattern": pattern})
        
#     pattern_list.append({"label": "GPE", "pattern": [{"LOWER": "bergen"}, {"lower": "op"}, {"lower": "zoom"}]})
#     pattern_list.append({"label": "GPE", "pattern": "bergen op zoom"})
    
    # Add drug and quantity rules to the model
    config = {
   "phrase_matcher_attr": None,
   "validate": True,
   "overwrite_ents": True,
   "ent_id_sep": "||",
    }
    ruler = nlp.add_pipe("entity_ruler", config=config)

    ruler.add_patterns(pattern_list)
```

```python
configure_spacy_model()
```

```python
nlp.to_disk('models/configured_spacy_model')
```
